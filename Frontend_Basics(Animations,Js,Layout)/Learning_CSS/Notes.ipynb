## **CSS Learning**

```CSS

h1, h2 {
	color: value;
}

```

Above targets all the h1 and h2. But `h1 h2` would only target `h2` that are inside a `h1` tag.

```CSS

p {
	color: purple!important;
}
```

In the above code `important` overrides every other styles for this tag.

```CSS
* {
	margin: 0;
	padding: 0;
	box-sizing: border-box;
}
```

The `*` is universal selector which is applied to every element in the page.

It is also called CSS reset.
